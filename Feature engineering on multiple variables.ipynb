```python
def one_hot_top_x(df,variable,top_labels):
    for label in top_labels:
        df[variable+'_'+label]=np.where(df[variable]==label,1,0)
        #df.variable would not have worked bcz input is 'X2'
```

```python
#function for findin top10
def top_10_x(variable):
    top=[x for x in data[variable].value_counts().sort_values(ascending=False).head(10).index]
    one_hot_top_x(data,variable,top)
```
